```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = "results.dat"
results = np.loadtxt(fname)
epoch = results[:,0]+1
train_elbo = results[:,1]
val_elbo = results[:,2]
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.title("(a) VAE loss over training process")
plt.plot(epoch, train_elbo, label="training")
plt.plot(epoch, val_elbo, label="validation")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()

plt.subplot(122)
plt.title("(b) VAE ELBO over training process")
plt.plot(epoch, (-train_elbo), label="training")
plt.plot(epoch, (-val_elbo), label="validation")
plt.xlabel("epoch")
plt.ylabel("ELBO")
plt.legend()

plt.tight_layout()
plt.show()
```
