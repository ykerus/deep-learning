```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname   = ["RNN_GRAD.dat","LSTM_GRAD.dat"]
labels  = ["RNN","LSTM"]

results = [np.loadtxt(f) for f in fname]
step    = [results[i][:,0]+1 for i in range(len(fname))]
norm    = [results[i][:,1] for i in range(len(fname))]
```

```python
plt.figure(figsize=(5,4))

plt.title("Gradients over time")
for i in range(len(fname)):
    p=plt.plot(step[i], norm[i], label = "{}".format(labels[i]))
#     p=plt.plot(step[i], acc[i], label = "{}, lr={}".format(labels[i],lr[i]))
#     plt.plot(step[i], acc[i], "--", color=p[0].get_color(),label = "test ({})".format(labels[i]))
plt.ylabel("gradient norm")
plt.xlabel("time step")

plt.legend(loc="right")

plt.tight_layout()
plt.show()
```
