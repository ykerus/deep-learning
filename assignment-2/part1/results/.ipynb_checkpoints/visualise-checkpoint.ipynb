```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname   = ["RNN_T05.dat","RNN_T09.dat","RNN_T13.dat","RNN_T17.dat","RNN_T21.dat","RNN_T25.dat",
            "RNN_T29.dat","RNN_T33.dat"]
labels  = ["T=5","T=9","T=13","T=17","T=21","T=25","T=29","T=33"]

# fname   = ["LSTM_T05.dat","LSTM_T09.dat","LSTM_T13.dat","LSTM_T17.dat","LSTM_T21.dat","LSTM_T25.dat",
#           "LSTM_T29.dat","LSTM_T33.dat"]
# labels  = ["T=5","T=9","T=13","T=17","T=21","T=25","T=29","T=33"]

# fname = ["RNN_T5.dat", "LSTM_T5.dat","RNN_T9.dat", "LSTM_T9.dat","RNN_T13.dat", "LSTM_T13.dat",
#         "RNN_T17.dat", "LSTM_T17.dat","RNN_T21.dat", "LSTM_T21.dat"]
# labels = ["RNN T5","LSTM T5","RNN T9","LSTM T9","RNN T13","LSTM T13","RNN T17","LSTM T17",
#          "RNN T21","LSTM T21"]

results = [np.loadtxt(f) for f in fname]
step    = [results[i][:,0] for i in range(len(fname))]
acc     = [results[i][:,1] for i in range(len(fname))]
loss    = [results[i][:,2] for i in range(len(fname))]
```

```python
lr = []
for i in range(len(fname)):
    f = open(fname[i], "r")
#     print("\n---- model {} -----".format(i+1))
    for i in range(9):
        info = f.readline()[1:-1]
#         print(info)
        lr.append(float(info[-5:])) if i == 6 else None
    f.close()
    
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.title("LSTM: loss over training process")
for i in range(len(fname)):
    p=plt.plot(step[i], loss[i], label = "{}".format(labels[i]))
#     p=plt.plot(step[i], loss[i], label = "{}, lr={}".format(labels[i],lr[i]))
#     plt.plot(step[i], loss[i],"--",color=p[0].get_color(),label="test ({})".format(labels[i]))
plt.ylabel("loss")
plt.xlabel("step")
# plt.xlim(None,150)
plt.legend()

plt.subplot(122)
plt.title("LSTM: accuracy over training process")
for i in range(len(fname)):
    p=plt.plot(step[i], acc[i], label = "{}".format(labels[i]))
#     p=plt.plot(step[i], acc[i], label = "{}, lr={}".format(labels[i],lr[i]))
#     plt.plot(step[i], acc[i], "--", color=p[0].get_color(),label = "test ({})".format(labels[i]))
plt.ylabel("accuracy")
plt.xlabel("step")
# plt.xlim(None,150)
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
