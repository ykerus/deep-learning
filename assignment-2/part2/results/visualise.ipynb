```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname   = ["FAIRY.dat"]
labels  = ["T"]

results = [np.loadtxt(f) for f in fname]
step    = [results[i][:,0] for i in range(len(fname))]
acc     = [results[i][:,1] for i in range(len(fname))]
loss    = [results[i][:,2] for i in range(len(fname))]
```
